# Introduction

> This project uses the TMDb Movie Dataset that contains more than 10,000 records of movies with details on title, budget, revenue, cast, directors, popularities, etc. The dataset allows to explore two key questions as follows: 
- The general trends are associated with movies over time?
- What are the common traits among the profitable films regarding genre, actors, directors, production firms, budget and runtime.

Which genres are most popular from year to year? 
What kinds of properties are associated with movies that have high revenues?

```python
# Use this cell to set up import statements for all of the packages that you plan to use
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Data Wrangling

> In the first section, the data will be loaded, examined and cleaned for the further analysis. Following steps will be carried out:
1. Understand the general property: df.info(), df.describe(), df.shape, df.head()
2. Drop duplicates
3. Drop unncessary columns 
4. Inspect data types
5. Determine cleaning steps


## General Properties

```python
df = pd.read_csv('tmdb-movies.csv')
df.info();
df.describe();
df.head(1);
f'The original dataset contains {df.shape[0]} entries and {df.shape[1]} columns' 
```

## Drop duplicates and n/a if necessary

```python
df.duplicated().sum() # remove 1 duplicates
df = df.drop_duplicates()
```

```python
df.isna().sum() 
# Missing data in several columns (e.g.cast, director, genres, production-companies), which do not impact on our analysis so no action is required. 
```

## Drop unnecessary columns
> - Only keep 13 relevant columns that answer the questions: 'popularity', 'budget', 'revenue', 'original_title',  'cast', 'director', 
    'runtime', 'genres', 'production_companies', 'release_date','vote_count', 'vote_average'
> - Remove Irrelevant columns: id, imdb_id, budget_adj, revenue_adj, homepage, tagline, keywords and overview.

```python
df.columns
df.drop(['id', 'imdb_id','homepage','tagline', 'keywords', 
         'overview','budget_adj', 'revenue_adj' ], 
        axis = 1, 
        inplace= True)
```

##  Inspect data, correct data types and clean data

```python
df.info() 
# wrong datatype: release_date => change from object to datetime
```

```python
df.describe()
# Zero value in budget, revenue and runtime that needs to be handle.
```

```python
df['release_date'] = pd.to_datetime(df['release_date'])
```

```python
zero_col = ['budget', 'revenue', 'runtime']
# Replace all 0 values by NaN using numpy
df[zero_col] = df[zero_col].replace(0,np.NAN)
df.dropna(subset = zero_col, inplace=True)
row, col = df.shape
f'There are {row-1} entries and {col} columns'
```

```python
# Double check the property of clean data. 
df.describe(); 
df.info()
```

```python
f'After cleaning, the clean dataset has {df.shape[0]} records of movies and {df.shape[1]} columns or properties'
```

# Exploratory Data Analysis
> After clean data is obtained, it's time to explore data by computing statistics and creating visualizations with the goal of addressing the RQs, looking at one variable at a time, and relationship among variables.

## General trends
It is to explore the movie distribution and the most profitable movie distribution overtime

### How many movies are profitably released yearly with profit over $100 million?

```python
df['profit'] = df['revenue'] - df['budget']
# Create a dataframe for only profitable movies 
df_profit = df[df['profit']>100000000]
f'{df_profit.shape[0]} out of {df.shape[0]} movies ({df_profit.shape[0]*100/df.shape[0]:.2f}%) generate a revenue of 100 million dollars or greater'
```

### What is the distribution of movies and profitable movie over time?

```python
df['release_year'].value_counts().sort_index().plot(figsize=(16,8),color='g',
                                                   label = 'Number of movies'); 
df_profit['release_year'].value_counts().sort_index().plot(figsize=(16,8), color = 'r',
                                                           label = 'Number of profitable movies')
plt.title('The distribution of movies released per year', fontsize=20)
plt.legend(fontsize=16)
plt.xlabel('Year', fontsize=16)
plt.ylabel('Number of movies',fontsize=16);
```

### What is the most profitable movie of all time

```python
most_prof_one = df[df['profit']>100000000].sort_values('budget',ascending = False).head(1)
f'The most profitable movie of all time is {most_prof_one.iloc[0,3]}'
```

### What are the average runtimes of the movie?

```python
df.groupby('release_year')['runtime'].mean().plot(kind='line',figsize=(16,8),color ='g')
plt.title('Average Runtime of moviews over years', fontsize=20)
plt.xlabel('Year', fontsize=16)
plt.ylabel('Average Runtime', fontsize=16);
```

### What is the total profit trend from year to year?

```python
df.groupby('release_year')['profit'].sum().plot(kind = 'line', figsize=(16,8), color = 'g');
plt.title('Profit trends over 50 years', fontsize = 20);
plt.xlabel('Year', fontsize =16 )
plt.ylabel('Profit ($)', fontsize=16);
```

### What is the average popularity in different budget range and revenue range?

```python
df['revenue_range'] = pd.cut(df['revenue'], df['revenue'].describe()[3:8],
                            labels = ['Low', 'Medium', 'High', 'Very High'])
df['revenue_range'].value_counts()
```

```python
df['budget_range']=pd.cut(df['budget'], df['budget'].describe()[3:8], 
                          labels = ['Low', 'Medium', 'High', 'Very High'])
df['budget_range'].value_counts()
```

```python
fig, (ax1, ax2) = plt.subplots(1,2)

df.groupby('revenue_range')['popularity'].mean().plot(kind='bar', ax = ax1, color = 'g', figsize=(16,8))
ax1.set_title('Average popularity base on revenue range')
ax1.xaxis.set_label_text('Revenue range', fontsize =12 )
ax1.yaxis.set_label_text('Popularity', fontsize =12 )
ax1.xaxis.set_tick_params(rotation = 45)

df.groupby('budget_range')['popularity'].mean().plot(kind='bar', ax = ax2, color = 'g'); 
ax2.set_title('Average popularity base on budget range')
ax2.xaxis.set_label_text('Budget range', fontsize =12 )
ax2.yaxis.set_label_text('Popularity', fontsize =12 )
ax2.xaxis.set_tick_params(rotation = 45)
```

### The most expensive and cheapest profitable movies of all time?

```python
most_exp_budget = df[df['profit']>100000000].sort_values('budget',ascending = True).head(5)
cheapest_budget = df[df['profit']>100000000].sort_values('budget',ascending = False).head(5)
print(f'The most expensive profitable movies of all times include: {most_exp_budget.iloc[0:4,3].tolist()}')
f'The cheapest profitable movies of all times include: {cheapest_budget.iloc[0:4,3].tolist()}'
```

## Properties of the most profitable movies?
What are the properties associated with the movies making more than 100 million dollars?
Properties include: actors, film producers, directors, average budget, average run-time

```python
#function which will take any column as argument from and keep its track
def calculate_count(column):
    data = df_profit[column].str.cat(sep = '|')  # Convert column to string and seperate it by '|'
    data = pd.Series(data.split('|'))      # Storing the values seperately in a Pandas series
    count = data.value_counts(ascending = False)
    return count
```

### Which types of genres are frequently associated with the profitable movies?

```python
genre_count = calculate_count('genres')
f'Genres frequently associated with the successful movies include: {genre_count[:5].index.tolist()}'
```

### Top actors are frequently associated with the profitable movies?

```python
actors_count = calculate_count('cast')
f'The top five actors have many successful movies include: {actors_count.head(5).index.tolist()}'
```

### Top directors are frequently associated with the profitable movies?

```python
director_count = calculate_count('director')
f'The top five directors have many successful movies include: {director_count.head(5).index.tolist()}'
```

### Top production firms are frequently associated with the profitable movies?

```python
prodfirm_count = calculate_count('production_companies')
f'The top five producers have many successful movies include: {prodfirm_count.head(5).index.tolist()}'
```

```python
fig = plt.figure(figsize=(16,8))
axes = fig.subplots(nrows=2, ncols=2)
axes[0,0].barh(genre_count.head(5).index, genre_count.head(5), color = 'g')
axes[0,0].set_title('Genres most frequently associated with the successful movies')
axes[0,0].xaxis.set_label_text('Number of the successful movies', fontsize =12 )

axes[0,1].barh(actors_count.head(5).index, actors_count.head(5), color = 'g')
axes[0,1].set_title('Actors most frequently associated with the successful movies')
axes[0,1].xaxis.set_label_text('Number of the successful movies', fontsize =12 )

axes[1,0].barh(prodfirm_count.head(5).index,prodfirm_count.head(5), color = 'g')
axes[1,0].set_title('Production firms most frequently associated with the successful movies')
axes[1,0].xaxis.set_label_text('Number of the successful movies', fontsize =12 )

axes[1,1].barh(director_count.head(5).index, director_count.head(5), color = 'g')
axes[1,1].set_title('Directors most frequently associated with the successful movies')
axes[1,1].xaxis.set_label_text('Number of the successful movies', fontsize =12 );
plt.tight_layout()
```

### What are the average budget, runtime and profits of the profitable movies?

```python
average_budget = df_profit['budget'].mean()
f'The average budget of a successful movie is ${average_budget/1e6:.2f} millions'
```

```python
average_runtime = df_profit['runtime'].mean()
f'The average runtime of the successful movie is {average_runtime/60:.2f} hour'
```

```python
revenue = df_profit['revenue'].mean()
f'The average revenue of the successful movie is ${revenue/1e6:.2f} millions'
```
